## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
# Lets check if there is any outliers using box plots
numerical_columns = df.select_dtypes("number").columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 5, i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# i will calculate the IQR for each numerical column
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df= df[~((df[numerical_columns] < lower_bound) | (df[numerical_columns] > upper_bound)).any(axis=1)]

# lets check if we have removed the outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 5, i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Impute the missing value using mode
df.fillna(df.mode().iloc[0], inplace=True)
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_enoder=LabelEncoder() # for all columns
df_encoder=df.copy()
for i in df_encoder.columns:
    if df_encoder[i].dtype=="object":
        df_encoder[i]=label_enoder.fit_transform(df[i])
df_encoder.reset_index(drop=True,inplace=True)
df_encoder.head()
```

```python
scaler = StandardScaler()
num=df.select_dtypes(include=[np.number]).columns
df_encoder[num] = scaler.fit_transform(df_encoder[num])
df_encoder
```

```python
pca = PCA(n_components=2)
pca_features = pca.fit_transform(df_encoder)

# Add PCA components to the dataframe
df_encoder['PCA1'] = pca_features[:, 0]
df_encoder['PCA2'] = pca_features[:, 1]
```

## Feature Selection
Select relevant features for clustering. Explain your choice of features.


```python
df_encoder= df_encoder[['PCA1', 'PCA2']]
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_encoder)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Lets train the model with the optimal number of clusters in my observations (k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoder['Cluster'] = kmeans.fit_predict(df_encoder)
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
silhouette_avg = silhouette_score(df_encoder[['PCA1', 'PCA2']], df_encoder['Cluster'])
print("Silhouette Score:", silhouette_avg)
```

## Plot the data points with their predicted cluster center

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_encoder, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label='Centroids')
plt.title('Clusters of Accidents')
plt.legend()
plt.show()
```
